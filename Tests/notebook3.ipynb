```python
# Dependencies
import pandas as pd
```

```python
vac = pd.read_csv('../Resources/covid19_vaccinations.csv')
vac
```

```python
vac2 = pd.read_csv('../Resources/b2020VAERSDATA.csv')
vac2
```

```python
# Merge two dataframes using an outer join
merge_df = pd.merge(vac, vac2, on="ABBR", how="outer")
merge_df
```

```python
# Remove the rows with missing data
clean_vac_df = merge_df.dropna(how="any")
clean_vac_df.count()
```

```python
clean_vac_df.head()
```

```python
clean_vac_df.tail()
```

```python
# Count for each state
state_counts = clean_vac_df["ABBR"].value_counts()
state_counts
```

```python
# Using GroupBy in order to separate the data into fields according to "state" values
grouped_vac_df = clean_vac_df.groupby(['ABBR'])

# The object returned is a "GroupBy" object and cannot be viewed normally...
print(grouped_vac_df)

# In order to be visualized, a data function must be used...
grouped_vac_df.count()
```
